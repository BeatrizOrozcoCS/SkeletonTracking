```python
from sklearn.linear_model import LinearRegression

# extract the input (X) and output (y) variables
X = electricCars[['mileage']]
y = electricCars['price']
# create a linear regression object and fit it to the data
reg = LinearRegression().fit(X, y)

regCoef = reg.coef_[0]
regIntercept = reg.intercept_

# print the slope (regression coefficient) and y-intercept of the LS line
print('Slope:', reg.coef_[0])
print('Y-intercept:', reg.intercept_)
```

```python
import statsmodels.api as sm

# extract the input (X) and output (y) variables
X = electricCars[['mileage']]
y = electricCars['price']

# add a constant to the input variables to fit an intercept
X = sm.add_constant(X)

# create a linear regression object and fit it to the data
reg = sm.OLS(y, X).fit()

# print the summary of the regression results
print(reg.summary())
```

```python
price = regCoef * 20000 + regIntercept
print("price of an electric car with mileage 20,000 miles is:", price)
```
